```python
import os
import re
import pandas as pd
```

```python
results_filenames = os.listdir(os.path.join('data', 'results'))
```

```python
results_filenames
```

```python
results_dfs = {}

for filename in results_filenames:
    archtecture = re.search(r'(?<=results_).*(?=.csv)', filename).group()
    results_dfs[archtecture] = pd.read_csv(os.path.join('data', 'results', filename))
```

```python
results_dfs.keys()
```

```python
results_dfs['SmallSqueezeNet'].head()
```

```python
results_dfs['ResNet10'].head()
```

```python
results_df = pd.concat(results_dfs.values(), ignore_index=True)
```

```python
results_df.tail()
```

```python
results_df['Architecture'].unique()
```

```python
classes = results_df['Classes'].unique()
```

```python
classes
```

```python
def getClassWithHigherValue(classes, column):
    for class_ in classes:
        yield results_df.iloc[
            results_df[results_df['Classes'] == class_][column].idxmax()]
```

```python
metabase_df = pd.DataFrame(
    getClassWithHigherValue(classes, 'Accuracy_avg')).reset_index(drop=True)
```

```python
metabase_df
```

```python
metabase_df.to_csv('metabase.csv', index=False)
```
