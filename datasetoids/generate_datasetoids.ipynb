```
import numpy as np
import pandas as pd
from seaborn import get_dataset_names, load_dataset
```

```
get_dataset_names()
```

```
titanic_df = load_dataset('titanic')
```

```
titanic_df.head()
```

```
def generate_datasetoids(dataset_df, symbolic_attributes):
    for symbolic_attribute in symbolic_attributes:
        X = dataset_df.drop(symbolic_attribute, axis=1).to_numpy()
        y = dataset_df[symbolic_attribute].to_numpy()
        yield X, y
```

```
datasetoids = list(generate_datasetoids(titanic_df, ['survived', 'age', 'class']))
```
